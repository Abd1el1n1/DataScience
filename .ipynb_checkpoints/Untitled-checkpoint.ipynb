```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
tickerSymbol = 'AMD'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2020-1-1', end='2020-4-3')

#see your data
tickerDf.to_csv('amd.csv', index=False)
```
