```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib as pyplot
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
UNH = 'UNH'

UNHData = yf.Ticker(UNH)

UNHDf = UNHData.history(period='1d',start='2019-1-1',end='2020-4-3')

#UNHDf.to_csv(str(UNH)+'.csv',index=False)

norm_UNH = UNHDf['High'].max()
vol_UNH = UNHDf['Volume'].max()
#fig, ax = plt.subplots()

#ax.legend(label='This is My Legend Title')

UNHDf['High'].div(norm_UNH).plot(xlim=['2020-1-1','2020-4-3'],label="se")
UNHDf['Volume'].div(vol_UNH).plot()

#UNHDf['High'].div(norm_UNH).plot(xlim=['2020-1-1','2020-4-3'],label="se")
#UNHDf['Volume'].div(vol_UNH).plot()
plt.legend()
```
